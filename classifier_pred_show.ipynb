```python
# load_model_sample.py
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import image
from keras.preprocessing.image import img_to_array
from keras.models import load_model
from keras.applications.imagenet_utils import decode_predictions



image_size = (244, 244)
path = "6_HUAWEI-MATE20.jpg"
img = keras.preprocessing.image.load_img(path,target_size=image_size)
img_array = keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
model = load_model("save_at_50.h5")
predictions = model.predict(img_array)
score = predictions[0]
print('score: '+str(score[0]))
print(
    "This image is %.2f percent BLURRY IMAGE  and %.2f percent Real SHARP IMAGE."
    % (100 * (1 - score), 100 * score)
)

plt.imshow(img)
plt.axis("off")
from PIL import Image
img = Image.open(path)
img.show()


```
